```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Reliance_Stock_Analysis/RILO - Copy.csv")
df.head()
```

```python
plt.figure(figsize=(14,10))
plt.plot(range(df.shape[0]) ,df['Close Price'])
plt.xticks(range(0,df.shape[0],100), df['Date'].loc[::100])
plt.title('reliance close price')
plt.xlabel('date')
plt.ylabel('Close Price')
plt.show()
```
